```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
Data = pd.read_csv("Energy Census and Economic Data US 2010-2014.csv")
# Data.count()
```

```python
Data_Reduced = Data[["State", "StateCodes", "Region", "Division", "TotalC2010", "TotalC2011", "TotalC2012", "TotalC2013", "TotalC2014", 
                     "TotalP2010", "TotalP2011", "TotalP2012", "TotalP2013", "TotalP2014",
                     "TotalE2010", "TotalE2011", "TotalE2012", "TotalE2013", "TotalE2014"]]
Data_Reduced = Data_Reduced[Data_Reduced.State != "United States"] 
Data_Reduced = Data_Reduced[Data_Reduced.State != "District of Columbia"]
Data_Reduced = Data_Reduced.sort_values(by=['TotalC2010'], ascending=False)
Data_Reduced.head()
```

```python
years = ["2010", "2011", "2012", "2013", "2014"]
years
```

```python
for year in years:

    # Plot Figure Size
    plt.figure(figsize=(8,3))

    # Plot Data
    plt.bar(Data_Reduced["StateCodes"], Data_Reduced["TotalC"+year], width=.9, align="center", label ="Consumption", color="red")
    plt.bar(Data_Reduced["StateCodes"], Data_Reduced["TotalP"+year], width=.4, align="center", label ="Production", color="orange")

    # Tick locations
    State = Data_Reduced["StateCodes"]
    x_axis = np.arange(len(State))
    tick_locations = [value for value in x_axis]
    plt.xticks(tick_locations, State, rotation=90, fontsize="x-small")

    # Plot Titles
    plt.title(year+" Energy Consumption vs. Production (State Level)", size="x-large")
    plt.xlabel("State")
    plt.ylabel("BTU's")

    # Margins
    plt.margins(x=.005)

    # Plot Legend
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Save Image
    SaveFile=("Output/"+year+"CP_State")
    plt.savefig(SaveFile)
```
