```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


import matplotlib.style as style 
# style.available
```

```python
plt.style.use('fast')
```

```python
Data = pd.read_csv("Energy Census and Economic Data US 2010-2014.csv")
Data.count()
```

```python
Data_Reduced = Data[["State", "StateCodes", "Region", "Division", "Coast", "Great Lakes", 
                     "TotalC2010", "TotalC2011", "TotalC2012", "TotalC2013", "TotalC2014", 
                     "TotalP2010", "TotalP2011", "TotalP2012", "TotalP2013", "TotalP2014",
                     "TotalE2010", "TotalE2011", "TotalE2012", "TotalE2013", "TotalE2014",
                     "TotalPrice2010", "TotalPrice2011", "TotalPrice2012", "TotalPrice2013", "TotalE2014",
                     "POPESTIMATE2010", "POPESTIMATE2011", "POPESTIMATE2012", "POPESTIMATE2013", "POPESTIMATE2014"]]
Data_Reduced = Data_Reduced[Data_Reduced.State != "United States"] 
Data_Reduced = Data_Reduced[Data_Reduced.State != "District of Columbia"]
Data_Reduced = Data_Reduced.sort_values(by=['TotalC2010'], ascending=False)
Data_Reduced.head()
```

```python
years = ["2010", "2011", "2012", "2013", "2014"]
years
```

```python
# Loop through each year creating plots 
for year in years:
    
    #fig = plt.figure(ncols=1, nrows=2)

    # Plot figure size
    # fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
    fig1, ax1 = plt.subplots(figsize=(9, 3))
    fig1, ax2 = plt.subplots(figsize=(9, 5))

    # ax1 ---------- ax1 ---------- ax1 ---------- ax1 ---------- ax1 #
    
    # Plot data
    ax1.bar(Data_Reduced["StateCodes"], Data_Reduced["TotalC"+year], width=.9, align="center", label ="Consumption")
    ax1.bar(Data_Reduced["StateCodes"], Data_Reduced["TotalP"+year], width=.4, align="center", label ="Production")

    # Rotate x-axis labels 90 degrees for ax1
    for tick in ax1.get_xticklabels():
        tick.set_rotation(90)
    
    # Plot titles and labels (as needed)
    ax1.set_title(year+"CY"+" Energy Consumption vs. Production\n(State Level)")
    ax1.set_ylabel("BTU's")
    
    # Adjust margins
    ax1.margins(x=.005)
    
    # Plot and format legend
    ax1.legend(loc="upper right", fontsize="small")
   
    # Diplay results
    plt.tight_layout
    plt.show()
    
    # ax2 ---------- ax2 ---------- ax2 ---------- ax2 ---------- ax2 #

    # Save Image
    # SaveFile=("Output/"+year+"CP_State")
    # fig1.savefig(SaveFile)
    
    # Diplay results
    plt.tight_layout
    plt.grid()
    plt.show()
```
