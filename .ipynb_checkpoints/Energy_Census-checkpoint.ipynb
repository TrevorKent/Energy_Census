```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("Energy Census and Economic Data US 2010-2014.csv")
Data.count()
```

```python
Data_Reduced = Data[["State", "Region", "Division", "TotalC2010", "TotalC2011", "TotalC2012", "TotalC2013", "TotalC2014", 
                     "TotalP2010", "TotalP2011", "TotalP2012", "TotalP2013", "TotalP2014",
                     "TotalE2010", "TotalE2011", "TotalE2012", "TotalE2013", "TotalE2014"]]
Data_Reduced = Data_Reduced[Data_Reduced.State != "United States"] 
Data_Reduced = Data_Reduced[Data_Reduced.State != "District of Columbia"]
Data_Reduced
```

```python
# Tick Locations
#tick_loactions = []
```

```python
# Plot Figure Size
plt.figure(figsize=(10,5))

# Plot Data
plt.bar(Data_Reduced["State"], Data_Reduced["TotalC2010"], width =.45, align="edge", label ="Consumption", color="red")
plt.bar(Data_Reduced["State"], Data_Reduced["TotalP2010"], width=-.45, align="edge", label ="Production", color="teal")

# Tick locations
State = Data_Reduced["State"]
x_axis = np.arange(len(State))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, State, rotation=90, fontsize="x-small")

# Plot Titles
plt.title("2010 Energy Consumption vs. Production (USA)")
plt.xlabel("State")
plt.ylabel("BTU's")

# Margins
plt.margins(x=.005)

# Plot Legend
plt.legend()
```
